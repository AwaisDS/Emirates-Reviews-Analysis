### Handling Missing Values

```python
df.Aircraft.value_counts()
```

```python
df['WiFi'].fillna(df['WiFi'].mean(), inplace=True)
df['Aircraft'].fillna('unknown', inplace=True)
```

```python
df.isna().sum()
```

```python
df.Status.unique()
```

```python
df['Status'].replace(np.nan, 'unknown', inplace=True)
df['Status'].replace('Not verified', 'Not Verified', inplace=True)
df['Status'].replace('❎ Not Verified', 'Not verified', inplace=True)
df['Status'].replace('❎ Unverified', 'Not verified', inplace=True)
```

```python
df.Status.unique()
```

```python
df.isna().sum()
```

```python
df['Seating Comfort'].fillna(df['Seating Comfort'].mean(), inplace=True)
df['Food Quality'].fillna(df['Food Quality'].mean(), inplace=True)
df['Entertainment'].fillna(df['Entertainment'].mean(), inplace=True)
df['Staff Service'].fillna(df['Staff Service'].mean(), inplace=True)
df['Ground Service'].fillna(df['Ground Service'].mean(), inplace=True)
```

```python
df.isna().sum()
```

## Exploratory Data Analysis 

### Top Drivers of Positive vs. Negative Reviews

```python
df.sample(3)
```

```python
from wordcloud import WordCloud

# Preparing the text
positive_text = " ".join(df[df["Overall Rating"] >= 8]["Review"].dropna()).lower()
negative_text = " ".join(df[df["Overall Rating"] <= 3]["Review"].dropna()).lower()

# Generating word clouds
positive_wc = WordCloud(width=800, height=400, background_color='white', colormap='Greens').generate(positive_text)
negative_wc = WordCloud(width=800, height=400, background_color='white', colormap='Reds').generate(negative_text)

fig, axs = plt.subplots(1, 2, figsize=(18, 8))

axs[0].imshow(positive_wc, interpolation='bilinear')
axs[0].set_title("Positive Reviews", fontsize=18)
axs[0].axis('off')

axs[1].imshow(negative_wc, interpolation='bilinear')
axs[1].set_title("Negative Reviews", fontsize=18)
axs[1].axis('off')

plt.tight_layout()
plt.show()
```

###  Sentiment vs. Star Rating

```python
from textblob import TextBlob

df["Sentiment Polarity"] = df["Review"].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

sns.scatterplot(data=df, x="Sentiment Polarity", y="Overall Rating", hue="Recommended")
plt.title("Sentiment vs Rating")
plt.show()
```

### Verified vs Not Verified

```python
print("Average Overall Rating by Status:")
df.groupby("Status")["Overall Rating"].mean()
```

```python
sns.boxplot(data=df, x="Status", y="Overall Rating")
plt.title("Verified vs Non-Verified Ratings")
plt.show()
```

### Ratings by Travel Type

```python
print("Average Overall Rating by Travel Type:")
df.groupby("Travel Type")["Overall Rating"].mean().sort_values(ascending=False)
```

### Sentiment Trend Over Time

```python
df["Date Published"] = pd.to_datetime(df["Date Published"], errors='coerce')
df["Year"] = df["Date Published"].dt.year

df.groupby("Year")[["Overall Rating", "Sentiment Polarity"]].mean().plot()
plt.title("Yearly Trend of Rating and Sentiment")
plt.show()
```

### Frequent Flyers' Reviews

```python
df["Frequent Flyer"] = df["Review"].str.contains("frequent", case=False, na=False)
df.groupby("Frequent Flyer")["Overall Rating"].mean()
```

### Value For Money

```python
df.groupby("Value for Money")["Overall Rating"].mean().plot(kind='bar', color='skyblue')
plt.title("Average Overall Rating vs Value for Money")
plt.xlabel("Value for Money Score")
plt.ylabel("Average Overall Rating")
plt.grid(True)
plt.show()
```
